```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
data = pd.read_csv("language.csv")
```

```python
data.head()
```

```python
# Find all unique values in the 'language' column
unique_languages = data['language'].unique()

print("Unique Languages:")
print(unique_languages)

print(f"\nTotal Unique Languages: {len(unique_languages)}")
```

```python
data.isnull().sum()
```

```python
data['language'].value_counts()
```

```python
data.dtypes
```

```python
x = np.array(data['Text'])
y = np.array(data['language'])
```

```python
print(x)
print("--"* 30)
print(y)
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x)
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size= 0.33, random_state = 42)
```

```python
X_train
```

```python
print(X_train)
```

```python
y_test
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
user = input("Enter a Text: ")
data = cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```

```python
import pickle

# Save CountVectorizer
with open("count_vectorizer.pkl", "wb") as f:
    pickle.dump(cv, f)

# Save Naive Bayes model
with open("language_detector.pkl", "wb") as f:
    pickle.dump(model, f)

print("Model Saved Successfully!")
```
